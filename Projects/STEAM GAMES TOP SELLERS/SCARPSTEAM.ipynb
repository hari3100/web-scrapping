```python
from bs4 import BeautifulSoup
import requests,openpyxl
import pandas as pd
```

```python
url = 'https://store.steampowered.com/search/?filter=globaltopsellers'
```

```python
source = requests.get(url)
source
```

```python
soup = BeautifulSoup(source.text,'html.parser')
soup
```

```python
chart = soup.find('div',id='search_resultsRows')
#print(chart)

indi_game = chart.find_all('a',class_='search_result_row ds_collapse_flag')
len(indi_game)
```

```python
data = []
for game in indi_game:
    
    name = game.find('span',class_='title').text
    
    date = game.find('div',class_='col search_released responsive_secondrow').text
    
    price = game.find('div',class_='col search_price_discount_combined responsive_secondrow').get_text(strip=True).split('₹')[-1]
    
    
    print(name,date,price)
    info = (name,date,price)
    data.append(info)
```

```python
scrap_data = pd.DataFrame(data,columns=['Name of the Game','Date of Release','Price of the Game(in ₹)'])
scrap_data
```
