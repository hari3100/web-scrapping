```python
from bs4 import BeautifulSoup
import requests
```

```python
# taking the source code of the desired website, after printing if the number is greater than 200 , dont scrap.
source  = requests.get("https://www.imdb.com/chart/top/")
source
```

```python
# bring that source html code in text form into the python with the help of beautifulsoup. 
soup = BeautifulSoup(source.text,'html.parser')
#print(soup)
# finding the needed tag with find()[which finds the 1st matching result] from the whole soup variable.
movies = soup.find('tbody',class_="lister-list")
#print(movies)
# finding the tbody tag with find()[which finds the 1st matching result] from the whole soup variable & then finding
# all the tr tag that had the details of the whole row with find_all()[which return all the matching results in a list].
movies = soup.find('tbody',class_="lister-list").find_all('tr')
# using len to se if we get all the 250 movies present.
print(len(movies))
```

```python
data=[]
# using loop to iterate from all the movies
for movie in movies:
    
    # every same as name but , .get_text() helps with str not callable error as this tag has multiple tags inside it
    # .get_text(strip=True) u will get the o/p in a single line it removes ant \n or tabs.
    # .split('.') is used to split the str where there is a . this will help us get the rank number only , as the split turns
    # the str to list , then we [0] take the index 0 .
    rank = movie.find('td','titleColumn').get_text(strip=True).split('.')[0]
    
    
    # finding the name of the movies, find()gives us the whole td tag that consist everything.Here we choose the name 
    # that was in the a tag hence we wrote .a[which gives the whole a tag] then 
    #.text[which will only provide the text in the a tag]
    name = movie.find('td',class_='titleColumn').a.text
    
    # every same the .strip() it's a str method that can be used to remove anything we want from the str we have, we removed ()
    year = movie.find('span','secondaryInfo').text.strip('()')
    
    
    
    rating = movie.find('td',class_='ratingColumn imdbRating').strong.text
  

    print(rank,name,year,rating)
    info = (rank,name,year,rating)
    data.append(info)
```

```python
import pandas as pd

IMDB_data = pd.DataFrame(data,columns = ['IMDB Ranking','Movie Name','Year of Release','IMDB Rating'])
IMDB_data
```
