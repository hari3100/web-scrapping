```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import openpyxl
```

```python
url = 'https://www.smallcase.com/discover/all?count=51&recentlyLaunched=false&sortBy=threeYearCagr&sortOrder=-1'
source = requests.get(url)
source
```

```python
soup = BeautifulSoup(source.text,'html.parser')
#soup
```

```python
allcases= soup.find('div',class_='AllSmallcases__sc-container__kVkpM flex ml24')
# allcases

# singlecase = allcases.find('a',class_='AllSmallcases__smallcasecard-link__2A7p_')
# singlecase
```

```python
indi_cases= soup.find('div',class_='AllSmallcases__sc-container__kVkpM flex ml24').find_all('a')
indi_cases

#cost= indi_cases.find('div',class_='SmallcaseCard__stats-container__3hR-P flex-apart SmallcaseCard__size-auto__NxnIC')
#cost
```

```python
data=[]
for case in indi_cases:
    
    #1
    name = case.find('h2',class_='SmallcaseCard__title__2M7E_ ellipsis mr8 SmallcaseCard__size-auto__NxnIC').text
    
    #2
    manager = case.find('span',class_='SmallcaseCard__hide-on-larger-size__1cICV SmallcaseCard__size-auto__NxnIC MetaData__managers-name__2RV67 text-light MetaData__text-13-mobile__2_k-R')
    manager = manager.text.strip('by')
    
    #3
    min_amt = case.find('div',class_='text-16 font-normal text-dark StatBox__value__2FWUJ mt8').text
    
    #4
    returns = case.find('div',class_='text-green text-16 font-normal text-dark font-medium StatBox__value__2FWUJ mt8').text
    
    #5
    risk = case.find('div',class_='SmallcaseCard__volatility-label__2B0b3 VolatilityLabel__volatility-tag-container__dcnic pt4 pb4 pl8 pr8 br-4 flex')
    risk = risk.text
    
    #6
    description = case.find('p',class_='SmallcaseCard__description__1H-JL text-14 text-normal lh-157 mb8 SmallcaseCard__size-auto__NxnIC')
    description = description.text
    
    print(name,manager,min_amt,returns,risk,description)
    info = (name,manager,min_amt,returns,risk,description)
    data.append(info)
```

```python
smallcase_data = pd.DataFrame(data,columns = ['Title','Managed By','Min. Investment','3Yrs Returns','Risk','Description'])
smallcase_data
```
